```python
import pprint, pickle
def load_pickles(name):
    pkl_file = open(name, 'rb')
    data = pickle.load(pkl_file)
    pkl_file.close()
    return data
```

```python
import pprint, pickle
import numpy as np
import matplotlib.pyplot as plt

'''
Performance metrics obtained by making 1000 Monte Carlo runs using the same Erdős–Rényi Model parameters (N=50, p=0.6) 
but different noise variance values
'''     
data1 = load_pickles('../saved_perf/erdos_50_0.6_1_perf.pkl')
data2 = load_pickles('../saved_perf/erdos_50_0.6_10_perf.pkl')
data3 = load_pickles('../saved_perf/erdos_50_0.6_100_perf.pkl')
data4 = load_pickles('../saved_perf/erdos_50_0.6_1000_perf.pkl')

x =np.arange(1,100,step=5)

data_plot1 = np.zeros((data1.shape[0],2))
for i in range(data1.shape[0]):
    data_plot1[i,:] = np.mean(data1[i,:,:],axis=0)

data_plot2 = np.zeros((data2.shape[0],2))
for i in range(data2.shape[0]):
    data_plot2[i,:] = np.mean(data2[i,:,:],axis=0)   

    
data_plot3 = np.zeros((data3.shape[0],2))
for i in range(data2.shape[0]):
    data_plot3[i,:] = np.mean(data3[i,:,:],axis=0)  
    
data_plot4 = np.zeros((data4.shape[0],2))
for i in range(data4.shape[0]):
    data_plot4[i,:] = np.mean(data4[i,:,:],axis=0)

fig=plt.figure()
plt.grid(alpha=0.5,linestyle='--')
ni=1
nx=15
plt.plot(x[ni:nx],data_plot1[ni:nx,0],linewidth=2)
plt.plot(x[ni:nx],data_plot1[ni:nx,1],'--',linewidth=2.5)
plt.plot(x[ni:nx],data_plot2[ni:nx,0],linewidth=2)
plt.plot(x[ni:nx],data_plot2[ni:nx,1],'--',linewidth=2.5)
plt.plot(x[ni:nx],data_plot3[ni:nx,0],linewidth=2)
plt.plot(x[ni:nx],data_plot3[ni:nx,1],'--',linewidth=2.5)
plt.plot(x[ni:nx],data_plot4[ni:nx,0],linewidth=2)
plt.plot(x[ni:nx],data_plot4[ni:nx,1],'--',linewidth=2.5)
plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
plt.title(r'$N=50, p=0.6$')
plt.ylabel("Accuracy (%)")
#plt.xlabel("Time-series dimension (n)")
plt.xlabel("Number of samples (n)")
plt.legend(["1 (SVM)","1 (CNN)","10 (SVM)","10 (CNN)", "100 (SVM)","100 (CNN)","1000 (SVM)","1000 (CNN)"], title=r'$\sigma$',loc='best')
fig.savefig("cnn+svm_accuracy_comparison_erdos_n50_p06.svg")
```
